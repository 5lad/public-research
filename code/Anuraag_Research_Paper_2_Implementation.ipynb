# Alpha Factor 2 Construction (Small Size)

```python
class Log_Market_Cap(CustomFactor):
    inputs = [morningstar.valuation.market_cap]
    window_length = 1

    def compute(self, today, assets, out, mc):
        out[:] = np.log(mc[-1])
    
        
```

```python
class Log_Market_Cap_Cubed(CustomFactor):
    inputs = [morningstar.valuation.market_cap]
    window_length = 1

    def compute(self, today, assets, out, mc):
        out[:] = np.log(mc[-1]**3)
        
        
```

```python
class Log_Stock_Price(CustomFactor):
    inputs=[USEquityPricing.close]
    window_length=1
    
    def compute(self,today,assets,out,sp):
        out[:] = np.log(sp[-1])
```

```python
class Log_Total_Last_Quarter_Assets(CustomFactor):
    inputs = [morningstar.balance_sheet.total_assets]
    window_length=63
    
    def compute(self,today,assets,out,tlqa):
        out[:] = np.log(tlqa[-1])
    
```

```python
class Log_TTM_Sales(CustomFactor):
    inputs = [morningstar.income_statement.total_revenue]
    window_length = 252
    
    def compute(self, today, assets, out, sps):
        out[:] = np.log(((sps[-1]) - (sps[0])) / (sps[0]))
```

```python
# This factor creates the synthetic S&P500
class SPY_proxy(CustomFactor):
    inputs = [morningstar.valuation.market_cap]
    window_length = 1
    
    def compute(self, today, assets, out, mc):
        out[:] = mc[-1]
```

### Pipeline for Factor 1

```python
def make_pipeline_for_factor_1():
    base_universe = Q1500US()
    pipe = Pipeline()
   
    ttm_sales_1_year_ago=ttm_sales(window_length=504,mask=base_universe)-ttm_sales(window_length=252,mask=base_universe)
    ttm_sales_today=ttm_sales(mask=base_universe)
    ratio_today=(ttm_sales_today-ttm_sales_1_year_ago)/(ttm_sales_1_year_ago)

    ttm_sales_1_year_3_months_ago=ttm_sales(window_length=567,mask=base_universe)-ttm_sales(window_length=315,mask=base_universe)
    ttm_sales_3_months_ago=ttm_sales(window_length=126,mask=base_universe)-ttm_sales(window_length=63,mask=base_universe)
    ratio_3_months_ago=(ttm_sales_3_months_ago-ttm_sales_1_year_3_months_ago)/(ttm_sales_1_year_3_months_ago)

    three_month_momentum=ratio_today/ratio_3_months_ago
    
    ttm_sales_1_year_ago=ttm_sales(window_length=504,mask=base_universe)-ttm_sales(window_length=252,mask=base_universe)
    ttm_sales_today=ttm_sales()
    ratio_today=(ttm_sales_today-ttm_sales_1_year_ago)/(ttm_sales_1_year_ago)


    ttm_sales_1_year_6_months_ago=ttm_sales(window_length=630,mask=base_universe)-ttm_sales(window_length=378,mask=base_universe)
    ttm_sales_6a_months_ago=ttm_sales(window_length=630,mask=base_universe)-ttm_sales(window_length=126,mask=base_universe)
    ratio_6_months_ago=(ttm_sales_6_months_ago-ttm_sales_1_year_6_months_ago)/(ttm_sales_1_year_6_months_ago)

    six_month_momentum=ratio_today/ratio_6_months_ago
    
    pipe.add(SPY_proxy(), 'SPY Proxy')
    pipe.add(three_month_momentum, "3-month momentum in traiing 12-month sales")
    pipe.add(six_month_momentum, "6-month momentum in trailing 12-month sales")
    #pipe.add(change_in_slope, "change in slope")
    
    return pipe
```

# Aggregation

### Used Gill Wassermann's aggregation methods/functions below for convenience

```python
def filter(x):
    if x <= -10:
        x = -10.0
    elif x >= 10:
        x = 10.0
    return x 

def standardize(df):
    
    # basic clean of dataset to remove infinite values
    df = df.replace([np.inf, -np.inf], np.nan)
    df = df.dropna()
    
    # need standardization params from synthetic S&P500
    df_SPY = df.sort(columns='SPY Proxy', ascending=False)

    # create separate dataframe for SPY
    # to store standardization values
    df_SPY = df_SPY.head(500)
    #print(df_SPY.head())
    
    # get dataframes into numpy array
    df_SPY = df_SPY.as_matrix()
    #print(df_SPY[1])
    
    
    # store index values
    index = df.index.values
    df = df.as_matrix()
    #print(df)
    
    df_standard = np.empty(df.shape[0])
   

    for col_SPY, col_full in zip(df_SPY.T, df.T):
        
        # summary stats for S&P500
        mu = np.mean(col_SPY)
        sigma = np.std(col_SPY)
        col_standard = np.array(((col_full - mu) / sigma)) 

        # create vectorized function (lambda equivalent)
        fltr = np.vectorize(filter)
        col_standard = (fltr(col_standard))
        
        # make range between -10 and 10
        col_standard = (col_standard / df.shape[1])
        
        # attach calculated values as new row in df_standard
        df_standard = np.vstack((df_standard, col_standard))
     
    # get rid of first entry (empty scores)
    df_standard = np.delete(df_standard,0,0)
    
    return (df_standard, index)

# Sum up and sort data
def composite_score(df, index):

    # sum up transformed data
    df_composite = df.sum(axis=0)
    
    # put into a pandas dataframe and connect numbers
    # to equities via reindexing
    df_composite = pd.Series(data=df_composite,index=index)
    
    # sort descending
    df_composite.sort(ascending=False)

    return df_composite

```

## Ranking Stocks Based on Factor 1 (Accelerated Sales)

```python
# compute the standardized values
results_standard_for_factor_1, index_for_factor_1 = standardize(results_for_factor_1)

# aggregate the scores
ranked_scores_for_factor_1 = composite_score(results_standard_for_factor_1, index_for_factor_1)

# print the final rankings
ranked_scores_for_factor_1
```
